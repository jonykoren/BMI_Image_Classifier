```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
pat = '/usr/bin/chromedriver'
```

```python
driver = webdriver.Chrome(pat)
```

```python
driver.get('https://height-weight-chart.com/')
```

```python
imgs = driver.find_elements_by_xpath('//img[@class="thumb"]')
```

```python
urls = []
for actual_image in imgs:
    if actual_image.get_attribute('src'):
        name = actual_image.get_attribute('src')
        name = name[34:]
        name = name[:7]
        url = 'https://height-weight-chart.com/'+str(name)+".html"
        urls.append(url)
        print(url)
```

```python
heights = []
weights = []
large_img_urls = []
for url in urls:
    driver.get(url)
    imgs = driver.find_elements_by_xpath('//img[@class="largepic"]')
    for actual_image in imgs:
        if actual_image.get_attribute('src'):
            large_img_url = actual_image.get_attribute('src')
            large_img_urls.append(large_img_url)
        text1 = driver.find_elements_by_xpath("//div[@class='main']/p[1]")
        for p in range(len(text1)):
            t_str = text1[p].text
            h = t_str.split()[0]
            w = t_str.split()[3]
            heights.append(h)
            weights.append(w)
    print(large_img_url, " , Weight = ",w," , Height = ",h,"\n")
        
```

```python
df = pd.DataFrame({"height":heights, "weight": weights, "img_url":large_img_urls})
df.to_csv("/home/jony/Desktop/bmi_dataset.csv")
```
