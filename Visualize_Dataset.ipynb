# Adding BMI feature

```python
df["bmi"] = df["weight"]/(df['height']/100*df['height']/100)
df
```

```python
# mean BMI
np.mean(df["bmi"])
```

```python
listy = []
lis = df["bmi"].tolist()
for i in lis:
    if (0<i<=18.5):
        print(i , "Underweight")
        listy.append("Underweight")
    if (18.5<i<=25):
        print(i , "Normal")
        listy.append("Normal")
    if (25<i<=30):
        print(i , "Overweight")
        listy.append("Overweight")
    if (30<i<=35):
        print(i , "Medium Obesity")
        listy.append("Medium Obesity")
    if (35<i<=135):
        print(i, "Super Obesity")
        listy.append("Super Obesity")
        
print(len(listy))
```

```python
df["meaning"] = listy
```

# torch dataset creation

```python
df = pd.read_csv("bmi_dataset3.csv")
```

```python
from PIL import Image as img2
```

```python
df2 = pd.DataFrame({"index":df["index"], "meaning":df["meaning"]})
df2
```

## Shuffle dataframe

```python
df2 = df2.reindex(np.random.permutation(df2.index))
train_subset = df2[0:878]
test_subset = df2[878:len(df2)]
```

```python
test_subset
```

```python
def rchop(thestring, ending):
    if thestring.endswith(ending):
        return thestring[:-len(ending)]
    return thestring
```

```python
pat = '/homedtic/ikoren/yalla/sel/data/'
TR = '/homedtic/ikoren/yalla/sel/train/'
TE = '/homedtic/ikoren/yalla/sel/test/'
range1 = train_subset["index"]
range2 = test_subset["index"]
for root, dirs, files in os.walk(pat):
    for file in files:
        p = os.path.join(root, file)
        print(p)
        if os.path.isfile(p):
            tmp = p[len(pat):]
            tmp2=rchop(tmp, '.png') 
            tmp2 = int(tmp2)
            
            if tmp2 in range1:
                class_tmp1 = train_subset["meaning"][tmp2]
                print("Train--> ", class_tmp1, '\n')
                im1 = img2.open(p)
                im1.save(TR+str(class_tmp1)+"/"+str(tmp2)+".png")
                
            if tmp2 in range2:
                class_tmp2 = test_subset["meaning"][tmp2]
                print("Test--> ", class_tmp2, '\n')
                im2 = img2.open(p)
                im2.save(TE+str(class_tmp2)+"/"+str(tmp2)+".png")
```

# Check data completeness

```python
def count_files(path):
    return sum([len(files) for _, _, files in os.walk(path)])

print("# of train images = ", count_files(TR)) # 878
print("# of test images = ", count_files(TE)) # 293

if count_files(TR)==878 and count_files(TE)==293:
    print("\n Data is good to go for training ANN")
```
