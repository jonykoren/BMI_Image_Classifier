```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/homedtic/ikoren/yalla/sel/bmi_dataset2.csv").drop(['Unnamed: 0'], axis=1)
df
```

# Adding BMI feature

```python
df["bmi"] = df["weight"]/(df['height']/100*df['height']/100)
df
```

```python
# mean BMI
np.mean(df["bmi"])
```

```python
listy = []
lis = df["bmi"].tolist()
for i in lis:
    if (0<i<=18.5):
        print(i , "Underweight")
        listy.append("Underweight")
    if (18.5<i<=25):
        print(i , "Normal")
        listy.append("Normal")
    if (25<i<=30):
        print(i , "Overweight")
        listy.append("Overweight")
    if (30<i<=35):
        print(i , "Medium Obesity")
        listy.append("Medium Obesity")
    if (35<i<=135):
        print(i, "Super Obesity")
        listy.append("Super Obesity")
        
print(len(listy))
```

```python
df["meaning"] = listy
```
